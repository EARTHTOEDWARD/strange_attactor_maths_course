# The Lorenz Attractor: The Butterfly Effect

In 1963, Edward Lorenz discovered the first strange attractor while studying weather prediction. His simple model of atmospheric convection revealed that deterministic systems could exhibit chaotic behavior.

The Lorenz system:
$$\frac{dx}{dt} = \sigma(y - x)$$
$$\frac{dy}{dt} = x(\rho - z) - y$$
$$\frac{dz}{dt} = xy - \beta z$$

This notebook will take you on a journey through one of the most beautiful objects in mathematics!

## 1. Understanding the Equations

The Lorenz system models fluid convection with three variables:
- **x**: Rate of convective motion
- **y**: Horizontal temperature variation 
- **z**: Vertical temperature variation

And three parameters:
- **√ (sigma)**: Prandtl number (ratio of viscosity to thermal diffusivity)
- **¡ (rho)**: Rayleigh number (temperature difference driving convection)
- **≤ (beta)**: Geometric factor

The classic parameter values (√=10, ¡=28, ≤=8/3) produce the famous butterfly attractor.

```python
# Create the Lorenz system with classic parameters
lorenz = LorenzSystem(sigma=10.0, rho=28.0, beta=8/3)

# Integrate from initial condition
x0 = np.array([1.0, 1.0, 1.0])
t, trajectory = lorenz.integrate(x0, t_span=(0, 50), n_points=5000)

print(f"Integrated {len(trajectory)} points over {t[-1]:.1f} time units")
print(f"Initial state: {x0}")
print(f"Final state: {trajectory[-1]}")
```

## 4. The Butterfly Effect

The term "butterfly effect" comes from the sensitive dependence on initial conditions - a butterfly flapping its wings in Brazil could theoretically cause a tornado in Texas weeks later.

```python
# Compare trajectories with tiny differences in initial conditions
lorenz = LorenzSystem()

# Three very similar initial conditions
initial_conditions = [
    np.array([1.0, 1.0, 1.0]),
    np.array([1.001, 1.0, 1.0]),  # 0.1% difference in x
    np.array([1.0, 1.001, 1.0])   # 0.1% difference in y
]

trajectories = []
for ic in initial_conditions:
    _, traj = lorenz.integrate(ic, t_span=(0, 40), n_points=4000)
    trajectories.append(traj)

# Plot sensitivity to initial conditions
labels = ['Original', 'xÄ + 0.001', 'yÄ + 0.001']
fig = plot_sensitivity(trajectories, labels=labels,
                      title="Butterfly Effect: Sensitivity to Initial Conditions")
plt.show()

# Calculate divergence over time
fig, ax = plt.subplots(figsize=(10, 6))

for i in range(1, len(trajectories)):
    distance = np.linalg.norm(trajectories[0] - trajectories[i], axis=1)
    ax.semilogy(t[:len(distance)], distance, linewidth=2, label=labels[i])

ax.set_xlabel('Time', fontsize=12)
ax.set_ylabel('Distance from Original Trajectory', fontsize=12)
ax.set_title('Exponential Divergence of Nearby Trajectories', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Initial separation: 0.001")
print(f"Final separation: {distance[-1]:.2f}")
print(f"Amplification: {distance[-1]/0.001:.0f}x")
```

## 5. Fixed Points and Stability

The Lorenz system has three fixed points where the derivatives are zero. Let's find and analyze them:

```python
def find_fixed_points(sigma, rho, beta):
    """Find fixed points of the Lorenz system."""
    # Fixed point equations:
    # √(y - x) = 0
    # x(¡ - z) - y = 0  
    # xy - ≤z = 0
    
    fixed_points = []
    
    # Origin is always a fixed point
    fixed_points.append(np.array([0, 0, 0]))
    
    # Non-trivial fixed points exist when ¡ > 1
    if rho > 1:
        x_star = np.sqrt(beta * (rho - 1))
        y_star = x_star
        z_star = rho - 1
        
        # Two symmetric fixed points
        fixed_points.append(np.array([x_star, y_star, z_star]))
        fixed_points.append(np.array([-x_star, -y_star, z_star]))
    
    return fixed_points

# Analyze stability through eigenvalues
def analyze_stability(lorenz, fixed_point):
    """Analyze stability of a fixed point."""
    jacobian = lorenz.jacobian(fixed_point)
    eigenvalues, _ = np.linalg.eig(jacobian)
    
    # Check real parts of eigenvalues
    real_parts = np.real(eigenvalues)
    
    if all(real_parts < 0):
        return "Stable (attracting)", eigenvalues
    elif all(real_parts > 0):
        return "Unstable (repelling)", eigenvalues
    else:
        return "Saddle point", eigenvalues

# Analyze for different parameter values
parameters = [
    (10, 5, 8/3, "Sub-critical (¡ < ¡c)"),
    (10, 28, 8/3, "Chaotic regime"),
]

for sigma, rho, beta, regime in parameters:
    print(f"\n{regime}: √={sigma}, ¡={rho}, ≤={beta:.3f}")
    print("=" * 50)
    
    lorenz = LorenzSystem(sigma, rho, beta)
    fixed_points = find_fixed_points(sigma, rho, beta)
    
    for i, fp in enumerate(fixed_points):
        stability, eigenvals = analyze_stability(lorenz, fp)
        print(f"\nFixed point {i+1}: {fp}")
        print(f"Stability: {stability}")
        print(f"Eigenvalues: {eigenvals}")

# Visualize fixed points and trajectories
fig = plt.figure(figsize=(12, 5))

for idx, (sigma, rho, beta, regime) in enumerate(parameters):
    ax = fig.add_subplot(1, 2, idx+1, projection='3d')
    
    # Integrate trajectory
    lorenz = LorenzSystem(sigma, rho, beta)
    _, trajectory = lorenz.integrate(np.array([1, 1, 1]), (0, 50), 5000)
    
    # Plot trajectory
    ax.plot(trajectory[:, 0], trajectory[:, 1], trajectory[:, 2],
            'b-', linewidth=0.5, alpha=0.6)
    
    # Plot fixed points
    fixed_points = find_fixed_points(sigma, rho, beta)
    for fp in fixed_points:
        ax.scatter(*fp, s=100, c='red', marker='o', edgecolors='black', linewidth=2)
    
    ax.set_title(regime)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()
```

## 8. Lorenz Map: Discrete Structure in Continuous Chaos

The Lorenz map reveals hidden order by plotting successive maxima of the z-coordinate:

```python
def extract_z_maxima(trajectory):
    """Extract local maxima of z-coordinate."""
    z = trajectory[:, 2]
    maxima = []
    
    for i in range(1, len(z)-1):
        if z[i] > z[i-1] and z[i] > z[i+1]:  # Local maximum
            maxima.append(z[i])
    
    return np.array(maxima)

# Generate long trajectory
lorenz = LorenzSystem()
_, trajectory = lorenz.integrate(np.array([1, 1, 1]), (0, 1000), n_points=100000)

# Extract z-maxima
z_maxima = extract_z_maxima(trajectory[10000:])  # Skip transient

# Plot Lorenz map
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Return map of maxima
ax1.scatter(z_maxima[:-1], z_maxima[1:], s=1, alpha=0.5)
ax1.set_xlabel('$z_n$ (current maximum)', fontsize=12)
ax1.set_ylabel('$z_{n+1}$ (next maximum)', fontsize=12)
ax1.set_title('Lorenz Map: Successive Z-Maxima', fontsize=14)
ax1.grid(True, alpha=0.3)

# Histogram of maxima
ax2.hist(z_maxima, bins=50, alpha=0.7, edgecolor='black')
ax2.set_xlabel('Z maximum value', fontsize=12)
ax2.set_ylabel('Count', fontsize=12)
ax2.set_title('Distribution of Z-Maxima', fontsize=14)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Found {len(z_maxima)} local maxima")
print(f"The tent-like shape reveals deterministic structure in the chaos!")
```

Let's simulate a simplified weather model:

```python
def weather_simulation(days=30, measurement_noise=0.1):
    """Simulate weather prediction with Lorenz dynamics."""
    
    # Scale time to represent days
    time_scale = 0.5  # Lorenz time units per day
    
    # Create two systems: actual and predicted
    actual_weather = LorenzSystem()
    predicted_weather = LorenzSystem()
    
    # Initial conditions (with measurement error)
    true_initial = np.array([1.0, 1.0, 1.0])
    measured_initial = true_initial + np.random.normal(0, measurement_noise, 3)
    
    # Integrate both systems
    t_span = (0, days * time_scale)
    n_points = days * 24  # Hourly data
    
    t, actual = actual_weather.integrate(true_initial, t_span, n_points)
    _, predicted = predicted_weather.integrate(measured_initial, t_span, n_points)
    
    # Convert to "temperature" (scaled z-coordinate)
    actual_temp = 15 + 10 * (actual[:, 2] - 25) / 20  # Celsius
    predicted_temp = 15 + 10 * (predicted[:, 2] - 25) / 20
    
    # Convert time to days
    time_days = t / time_scale
    
    # Calculate prediction error
    error = np.abs(actual_temp - predicted_temp)
    
    # Plot results
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
    
    # Temperature forecast
    ax1.plot(time_days, actual_temp, 'b-', linewidth=2, label='Actual')
    ax1.plot(time_days, predicted_temp, 'r--', linewidth=2, label='Predicted')
    ax1.set_ylabel('Temperature (∞C)', fontsize=12)
    ax1.set_title('Weather Forecast vs Reality', fontsize=14)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Prediction error
    ax2.plot(time_days, error, 'g-', linewidth=2)
    ax2.axhline(y=2, color='orange', linestyle='--', label='Acceptable error')
    ax2.set_xlabel('Days', fontsize=12)
    ax2.set_ylabel('Prediction Error (∞C)', fontsize=12)
    ax2.set_title('Forecast Accuracy Degradation', fontsize=14)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Find prediction horizon
    reliable_days = np.where(error > 2)[0]
    if len(reliable_days) > 0:
        horizon = time_days[reliable_days[0]]
        print(f"Reliable prediction horizon: {horizon:.1f} days")
    else:
        print(f"Predictions remain reliable for entire {days}-day period")
    
    print(f"Initial measurement error: {measurement_noise*10:.1f}∞C")
    print(f"Maximum prediction error: {error.max():.1f}∞C")

# Interactive weather predictor
interact(weather_simulation,
         days=IntSlider(min=5, max=60, step=5, value=30, description='Forecast days:'),
         measurement_noise=FloatSlider(min=0.01, max=0.5, step=0.01, value=0.1,
                                      description='Measurement error:'));
```

### Exercise 1: Lyapunov Exponents
Calculate the Lyapunov exponents for the Lorenz system to quantify chaos:

```python
def calculate_lyapunov_spectrum(lorenz, x0, t_final=1000, dt=0.01):
    """Calculate Lyapunov exponents using QR decomposition method."""
    # Your implementation here
    # This is a complex calculation - here's a simplified version
    
    from scipy.integrate import odeint
    
    def lorenz_extended(state, t):
        """Extended system including variational equations."""
        # First 3 components: original system
        x, y, z = state[:3]
        
        # Lorenz dynamics
        dx = lorenz.sigma * (y - x)
        dy = x * (lorenz.rho - z) - y  
        dz = x * y - lorenz.beta * z
        
        # Reshape tangent vectors (3x3 matrix stored as 9-vector)
        Phi = state[3:].reshape((3, 3))
        
        # Jacobian
        J = lorenz.jacobian(np.array([x, y, z]))
        
        # Variational equation: dPhi/dt = J * Phi
        dPhi = np.dot(J, Phi).flatten()
        
        return np.concatenate([[dx, dy, dz], dPhi])
    
    # Initial conditions: state + identity matrix
    initial = np.concatenate([x0, np.eye(3).flatten()])
    
    # Time points
    t = np.arange(0, t_final, dt)
    
    # Storage for Lyapunov sums
    lyap_sums = np.zeros(3)
    
    # Current state
    state = initial
    
    print("Calculating Lyapunov exponents...")
    
    # Integrate and orthonormalize periodically
    for i in range(1, len(t)):
        # Integrate one step
        state = odeint(lorenz_extended, state, [t[i-1], t[i]])[1]
        
        # Extract and orthonormalize tangent vectors
        Phi = state[3:].reshape((3, 3))
        Q, R = np.linalg.qr(Phi)
        
        # Accumulate growth rates
        lyap_sums += np.log(np.abs(np.diag(R)))
        
        # Replace tangent vectors with orthonormal ones
        state[3:] = Q.flatten()
        
        # Progress indicator
        if i % 1000 == 0:
            print(f"Progress: {i/len(t)*100:.1f}%")
    
    # Average to get exponents
    lyapunov_exponents = lyap_sums / (t_final)
    
    return np.sort(lyapunov_exponents)[::-1]  # Sort descending

# Calculate for classic parameters
lorenz = LorenzSystem()
lyap_exp = calculate_lyapunov_spectrum(lorenz, np.array([1, 1, 1]), t_final=100)

print("\nLyapunov Exponents:")
print(f"ªÅ = {lyap_exp[0]:.3f} (positive = chaos)")
print(f"ªÇ = {lyap_exp[1]:.3f} (near zero)")
print(f"ªÉ = {lyap_exp[2]:.3f} (negative = contraction)")
print(f"\nSum: {sum(lyap_exp):.3f} (should be negative for dissipative system)")
print(f"Kaplan-Yorke dimension: {2 - lyap_exp[0]/lyap_exp[2]:.3f}")
```

### Exercise 2: Control and Synchronization
Implement chaos synchronization between two Lorenz systems:

```python
def chaos_synchronization(coupling_strength=0.1, time=50):
    """Demonstrate synchronization of chaotic systems."""
    
    # Create master and slave systems
    master = LorenzSystem()
    
    # Different initial conditions
    master_ic = np.array([1, 1, 1])
    slave_ic = np.array([10, 10, 10])
    
    # Time array
    dt = 0.01
    t = np.arange(0, time, dt)
    n = len(t)
    
    # Storage
    master_traj = np.zeros((n, 3))
    slave_traj = np.zeros((n, 3))
    
    master_traj[0] = master_ic
    slave_traj[0] = slave_ic
    
    # Euler integration with coupling
    for i in range(1, n):
        # Master system (uncoupled)
        master_deriv = master.derivatives(master_traj[i-1])
        master_traj[i] = master_traj[i-1] + dt * master_deriv
        
        # Slave system with coupling
        slave_deriv = master.derivatives(slave_traj[i-1])
        
        # Add coupling term (only to x-component)
        coupling = coupling_strength * (master_traj[i-1, 0] - slave_traj[i-1, 0])
        slave_deriv[0] += coupling
        
        slave_traj[i] = slave_traj[i-1] + dt * slave_deriv
    
    # Calculate synchronization error
    sync_error = np.linalg.norm(master_traj - slave_traj, axis=1)
    
    # Plot results
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    
    # 3D trajectories
    ax = axes[0, 0] = plt.subplot(221, projection='3d')
    ax.plot(master_traj[:, 0], master_traj[:, 1], master_traj[:, 2],
            'b-', linewidth=0.5, alpha=0.6, label='Master')
    ax.plot(slave_traj[:, 0], slave_traj[:, 1], slave_traj[:, 2],
            'r-', linewidth=0.5, alpha=0.6, label='Slave')
    ax.set_title(f'Coupled Systems (k={coupling_strength})')
    ax.legend()
    
    # X-component comparison
    axes[0, 1].plot(t, master_traj[:, 0], 'b-', linewidth=1, label='Master')
    axes[0, 1].plot(t, slave_traj[:, 0], 'r--', linewidth=1, label='Slave')
    axes[0, 1].set_title('X-component')
    axes[0, 1].set_xlabel('Time')
    axes[0, 1].legend()
    
    # Synchronization error
    axes[1, 0].semilogy(t, sync_error, 'g-', linewidth=2)
    axes[1, 0].set_title('Synchronization Error')
    axes[1, 0].set_xlabel('Time')
    axes[1, 0].set_ylabel('||Master - Slave||')
    axes[1, 0].grid(True, alpha=0.3)
    
    # Phase difference
    axes[1, 1].plot(master_traj[:, 0], slave_traj[:, 0], 'k.', markersize=0.5, alpha=0.5)
    axes[1, 1].plot([-20, 20], [-20, 20], 'r--', label='Perfect sync')
    axes[1, 1].set_title('Master vs Slave (X)')
    axes[1, 1].set_xlabel('Master X')
    axes[1, 1].set_ylabel('Slave X')
    axes[1, 1].legend()
    axes[1, 1].set_aspect('equal')
    
    plt.tight_layout()
    plt.show()
    
    if sync_error[-1] < 0.1:
        print(f" Systems synchronized! Final error: {sync_error[-1]:.2e}")
    else:
        print(f" Systems not synchronized. Final error: {sync_error[-1]:.2f}")
        print("  Try increasing coupling strength.")

# Test synchronization
interact(chaos_synchronization,
         coupling_strength=FloatSlider(min=0, max=1, step=0.01, value=0.1,
                                      description='Coupling k:'),
         time=IntSlider(min=10, max=100, step=10, value=50,
                       description='Time:'));
```
